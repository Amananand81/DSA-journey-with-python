* store and Print all elements using a loop.

* Print the elements in reverse order.

```python
arr = [10, 20, 30, 40, 50]
for i in range(len(arr)-1,-1,-1):
    print(arr[i])
```

* sum of array items


```python
arr = [12, 8, 25, 10, 15]
sum=0
for i in range (len(arr)):
    sum=sum+arr[i]
print(sum)
```

* Find the largest element.

```python
arr = [12, 8, 25, 10, 15]
largest=arr[0]
for i in range(1,len(arr),1):
    if arr[i]>largest:
        largest=arr[i]
print(largest)
```

* find the smallest element 


```python
arr = [12, 8, 25, 10, 15]
smallest=arr[0]
for i in range(1,len(arr),1):
    if (arr[i]<smallest):
        smallest=arr[i]
print("smallest element: ",smallest)
```

* Count how many even numbers and how many odd numbers are present.

```python
arr = [12, 8, 25, 10, 15, 8, 25, 8]
countEven=0
countOdd=0
for i in range(0,len(arr),1):
    if(arr[i]%2==0):
        countEven+=1
    else:
        countOdd+=1
print("even: ",countEven)
print("odd: ",countOdd)
```
