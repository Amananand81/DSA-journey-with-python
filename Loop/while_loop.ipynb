* Write a Python program to calculate the sum of numbers from 1 to 100 using a while loop.

```python
i=1
sum=0
while i<=100:
    sum=sum+i
    i=i+1
print(sum)
```

* Digit extraction (very important for DSA)


```python
n=int(input("enter number"))
rev=0
while n!=0:
    num=n%10
    rev=rev*10+num
    n=n//10
print(rev)
```

* Palindrome/Reverse Number (Boss Challenge)

```python
n=12321
check=n
rev=0
while n!=0:
    num=n%10
    rev=rev*10+num
    n=n//10
print(rev)
if(rev==check):
    print("palindrome number")
else:
    print("not palindrome number")
```
